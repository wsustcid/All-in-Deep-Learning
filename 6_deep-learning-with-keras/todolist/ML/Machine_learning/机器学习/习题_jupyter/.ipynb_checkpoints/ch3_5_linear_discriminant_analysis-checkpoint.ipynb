# 3.5实现线性判别分析
编程实现线性判别分析，并给出西瓜数据集3.0alpha上的结果

```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
def get_x_y(x,y):
    return x[np.where(y==0),0], x[np.where(y==0),1], x[np.where(y==1),0], x[np.where(y==1),1]

def plot_boundary(x,y,x_train, y_train, x_test, y_test, weight):
    xx, yy = np.meshgrid(np.arange(min(x[:,0])-0.1, max(x[:,0])+0.1, 0.01),
                         np.arange(min(x[:,1])-0.1, max(x[:,1])+0.1, 0.01))
    zz = weight[0]*xx + weight[1]*yy
    f1 = plt.figure(1)   
    plt.title('watermelon_3a')  
    plt.xlabel('density')  
    plt.ylabel('ratio_sugar')  
    plt.contourf(xx,yy,zz)  
    plt_x0, plt_y0, plt_x1, plt_y1 = get_x_y(x_train, y_train)
    plt_xt0, plt_yt0, plt_xt1, plt_yt1 = get_x_y(x_test, y_test)
    plt.scatter(plt_x0, plt_y0, marker = 'o', color = 'red', s=30, label = 'train_bad')
    plt.scatter(plt_x1, plt_y1, marker = 'o', color = 'g', s=30, label = 'train_good')
    plt.scatter(plt_xt0, plt_yt0, marker = 'x', color = 'red', s=30, label = 'test_bad')
    plt.scatter(plt_xt1, plt_yt1, marker = 'x', color = 'g', s=30, label = 'test_good')
    plt.legend(loc = 'upper right')
    plt.show()
```

```python
dataset = np.loadtxt('data/watermelon_3_0_alpha.txt',delimiter=',')
x = dataset[:,1:3]
num_features = x.shape[1]
# x = np.insert(x,2,values=1,axis=1)
# x = x
y = dataset[:, 3]
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.5, random_state=0)
```

### 自己编写的LDA的结果

```python
weight = np.ones([2,1])
x0 = x_train[np.where(y_train==0)].T
x1 = x_train[np.where(y_train==1)].T
miu0 = np.mean(x0, axis = 1)
miu1 = np.mean(x1, axis = 1)
var0 = np.zeros([2,2])
var1 = np.zeros([2,2])
for i in range(len(x0)):
    var0 += np.dot(np.reshape((x0[:,i]-miu0),[2,1]), np.reshape((x0[:,i]-miu0).T,[1,2]))
for i in range(len(x1)):
    var1 += np.dot(np.reshape((x1[:,i]-miu1),[2,1]), np.reshape((x1[:,i]-miu1).T,[1,2]))

sw = var0 + var1
sw_inv = linalg.inv(sw)
weight = np.dot(sw_inv, np.reshape((miu0-miu1),[2,1]))

print(weight)

plot_boundary(x,y,x_train, y_train, x_test, y_test, np.ravel(weight))
```

### 调用sklearn的结果

```python
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
print(lda.coef_)
plot_boundary(x, y, x_train, y_train, x_test, y_test, np.ravel(lda.coef_))
```
